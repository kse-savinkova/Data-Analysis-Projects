## Задача 2

Определить количество обзоров и среднюю оценку для каждой книги

```python
query2 = '''
SELECT book_id,
       title,
       COUNT(DISTINCT review_id) AS count_reviews,
       ROUND(AVG(rating), 2) AS avg_rating
FROM books
LEFT JOIN reviews USING (book_id)
LEFT JOIN ratings USING (book_id)
GROUP BY book_id
ORDER BY count_reviews DESC;
'''

query2 = pd.io.sql.read_sql(sql=text(query2), con = con)
display(query2)
```

**Вывод:** Данные выведены в порядке убывания количества пользовательских обзоров, так можно увидеть, что не все книги имеют пользовательские обзоры. Вместе с тем, если вывести данные в порядке возрастания средней оценки книги, то мы увидим что все книги имеют оценки.

```python
display(query2.sort_values(by='avg_rating'))
```

## Задача 3

Определить издательство, которое выпустило наибольшее число книг (исключая брошюры < 50 страниц)

`Вариант 1 (если у нас только 1 издательство с максимальным количеством книг можно использовать LIMIT, так запрос выглядит локоничнее)`

```python

query3 = '''
SELECT publisher,
       COUNT(book_id)
FROM books 
JOIN publishers USING(publisher_id)
WHERE num_pages >= 50
GROUP BY publisher
ORDER BY count DESC
LIMIT 1;
'''

query3 = pd.io.sql.read_sql(sql=text(query3), con = con)
#display(query3)

print(f'Издательство-лидер по количеству выпущенных книг: {query3.iloc[0, 0]}')
print(f'Количество выпущенных книг: {query3.iloc[0, 1]}')
```

**Вывод:** Издательство которые выпустило больше всего книг Penguin Books.

`Вариант 2 (если бы было несколько издательств с максимальным количеством книг то следовало бы использовать MAX)`

```python
query3_1 = '''
WITH

--таблица содержащая название издательства и количество выпущенных им книг
a AS
(SELECT publisher,
        COUNT(book_id) AS count_books
FROM books 
JOIN publishers USING(publisher_id)
WHERE num_pages >= 50
GROUP BY publisher
)

--запрос который выбирает издательство(а) с максимальным количеством выпущенных книг
SELECT publisher,
       count_books
FROM a
WHERE count_books IN (SELECT MAX(count_books)
                      FROM a)
'''

query3_1 = pd.io.sql.read_sql(sql=text(query3_1), con = con)
display(query3_1)

print(f'Издательство-лидер по количеству выпущенных книг: {query3_1.iloc[0, 0]}')
print(f'Количество выпущенных книг: {query3_1.iloc[0, 1]}')
```

## Задача 4

Определить автора с самой высокой средней оценкой книг (книги с 50 и более оценками)

`Как я поняла задачу:
Определить автора с самой высокой средней оценкой средних оценок его книг. То есть сначала для каждого автора определяем среднюю оценку каждой его книги (ведь у каждой книги может быть несколько оценок) и уже потом находим среднюю оценку всех книг (то есть среднюю оценку средних оценок всех книг 1 автора).
При этом нас интересуют только книги с количеством оценок >=50.`

```python
query4 = '''
WITH
--таблица которая содержит авторов, их книги с количеством оценок >= 50 и среднюю оценку каждой такой книги автора
avg_rat_book AS
(SELECT author_id,
       author,
       book_id,
       avg(rating) AS avg_rat_book,
       count(rating) AS count_ratings
FROM books AS b
JOIN authors AS a USING(author_id)
JOIN ratings AS r USING(book_id)
GROUP BY author_id, author, book_id
HAVING count(rating) >= 50
ORDER BY author_id)

--таблица которая содержит автора и среднюю оценку средних оценок его книг
SELECT a.author,
       round(AVG(a.avg_rat_book), 2) AS avg_rating
FROM avg_rat_book AS a
GROUP BY a.author
ORDER BY avg_rating DESC
LIMIT 1;
'''

query4 = pd.io.sql.read_sql(sql=text(query4), con = con)
#display(query)
print(f'Автор с самой высокой средней оценкой книг: {query4.iloc[0, 0]}')
print(f'Средняя оценка книг: {query4.iloc[0, 1]}')
```

**Вывод:** Автор с самой высокой средней оценкой книг: J.K. Rowling/Mary GrandPré

## Задача 5

Определить среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
query5 = '''
SELECT ROUND(AVG(count_reviews), 2)
FROM (
--таблица которая содержит пользователей, количество оценок поставленных ими 
--и количество обзоров каждого пользователя
    SELECT username,
           COUNT(DISTINCT rating_id) AS count_ratings,
           COUNT(DISTINCT review_id) AS count_reviews
    FROM ratings
    JOIN reviews USING(username)
    GROUP BY username
    HAVING COUNT(DISTINCT rating_id) > 48) AS a
'''

query5 = pd.io.sql.read_sql(sql=text(query5), con = con)
#display(query5)
print(f'Среднее количество обзоров от пользователей которые поставили больше \
48 оценок: {query5.iloc[0, 0]} обзоров')
```

**Вывод:** Пользователи, которые поставили больше 48 оценок, пишут в среднем около 24 обзоров на книги.
